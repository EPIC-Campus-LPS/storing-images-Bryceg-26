7. We can also flip an image 90 degrees left or right by modifying the pixels of the image. Create a new image that is a 90 degree rotation counterclockwise of the original.

```python
from PIL import Image

original_image = Image.open("moo.webp") #determining what the image is called

width, height = original_image.size #this is finding the height and width of the image

rotated_image = Image.new("RGB", (height, width)) #creates a blank image of the sane size rotated

original_pixels = original_image.load() #puts the color back
rotated_pixels = rotated_image.load() #rotates the color pixels to fit the image

for x in range(width): #uses for loops to find each pixels color of the upright photo and rotates it to it's new home
    for y in range(height):
        rotated_pixels[y, width - 1 - x] = original_pixels[x, y]

t = plt.imshow(rotated_image) #displays the image
```
